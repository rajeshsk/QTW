```python
 def strip_head(text):
        tmp = []
        flag = False
        for i in text:
            if i=='':
                flag=True
            if flag==True:
                tmp.append(i)
        return tmp 
            
```

```python
# remove special characters 
def remove_special_character(text,remove_digits=False):
    pattern = r'[^a-zA-z0-9]' if not remove_digits else r'[^a-zA-z\s]'
    text = re.sub(pattern,' ',text)
    return text
```

```python
def remove_Square_bracks(text):
    pattern  = r'\[.*?\]'
    ntext = re.sub(pattern,'',text)
    return ntext
```

```python
df = []
example = os.listdir("./easy_ham//")
for i in example:
    
    with open("./easy_ham/"+i,"r",encoding='windows-1252') as f:
        data = [i.strip() for i in f]
        data = strip_head(data)
        
        data_join = ' '.join(data)
        data_strip_links = re.sub(r'http\S+', '',data_join) # remove https links 
        data_no_special_characters = remove_special_character(data_strip_links) # remove speical characters 
        clean_text = remove_Square_bracks(data_no_special_characters) # remove [digit],[] etc 
        clean_text = clean_text.replace('_',' ') # move lines 
        
        
        print(clean_text)
#df.append(data_join)
        
 
```
