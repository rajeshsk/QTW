```python
har_ham_list=[]
hard_ham = os.listdir('./hard_ham/')

for i in hard_ham:
    
    with open("./hard_ham/"+i,"r",encoding='windows-1252') as f:
        data = [i.strip() for i in f]
        data = strip_head(data)
        
        data_join = ' '.join(data)
        data_strip_links = re.sub(r'http\S+', '',data_join) # remove https links 
        data_no_special_characters = remove_special_character(data_strip_links) # remove speical characters 
        clean_text = remove_Square_bracks(data_no_special_characters) # remove [digit],[] etc 
        
        clean_text = clean_text.replace('_',' ') # move lines 
        clean_text=re.sub(' \d+','',clean_text) # remove unnecessary digits. 
        
        
        #print(clean_text)
    har_ham_list.append(clean_text)
```

### Spam

```python
df_spam = []
spam_list= os.listdir('./spam/')
for i in spam_list:
    
    with open('./spam/'+i,"r",errors="ignore") as f:
        data = [i.strip() for i in f]
        data = strip_head(data)
        
        data_join = ' '.join(data)
        data_strip_links = re.sub(r'http\S+', '',data_join) # remove https links 
        data_no_special_characters = remove_special_character(data_strip_links) # remove speical characters 
        clean_text = remove_Square_bracks(data_no_special_characters) # remove [digit],[] etc 
        clean_text = clean_text.replace('_',' ') # move lines 
        
        
        #print(clean_text)
    df_spam.append(clean_text)
```
